# (66391) Moshup propagation test

```python
import grss
prop = grss.prop
fit = grss.fit
```

```python
import numpy as np
np.set_printoptions(precision=40, linewidth=np.inf)
import matplotlib.pyplot as plt
```

```python
body_id = '66391'
init_sol, init_cov, nongrav_info = fit.get_sbdb_info(body_id)
```

```python
e = init_sol['e']
q = init_sol['q']
tp = init_sol['tp']
om = init_sol['om']*np.pi/180.0
w = init_sol['w']*np.pi/180.0
i = init_sol['i']*np.pi/180.0

cometary_elements = [e, q, tp, om, w, i]
ng_params = prop.NongravParamaters()
ng_params.a1 = init_sol.get('a1', 0.0)
ng_params.a2 = init_sol.get('a2', 0.0)
ng_params.a3 = init_sol.get('a3', 0.0)
```

```python
de_kernel = 441
de_kernel_path = grss.utils.default_kernel_path(de_kernel)
t0 = init_sol['t']
daysInYear = 365.25
numYears = 10
numDays = numYears * daysInYear
tf = t0 + numDays
prop_sim = prop.propSimulation("(66391) Moshup propagation test", t0, de_kernel, de_kernel_path)
```

```python
t_temp = t0
t_eval = []
while t_temp <= tf:
    t_eval.append(t_temp)
    t_temp += 1.0

eval_apparent_state = False
t_eval_utc = False
converged_light_time = False
prop_sim.set_integration_parameters(tf, t_eval, t_eval_utc, eval_apparent_state, converged_light_time)
```

```python
moshup = prop.IntegBody("(66391) Moshup", t0, 0.0, 0.0, cometary_elements, ng_params)
prop_sim.add_integ_body(moshup)
```

```python
prop_sim.integrate()
```

```python
arr = np.array(prop_sim.xIntegEval)
# plot xy and xz projections of the orbit in a 2x1 subplot
plt.figure(figsize=(6, 6), dpi=150)
axs = plt.gca()
prop.plot_solar_system(axs, xy_plane=True, alpha=0.5)
axs.plot(arr[:,0], arr[:,1], '-', lw=0.2)
axs.set_xlabel("x [AU]")
axs.set_ylabel("y [AU]")
axs.set_aspect('equal')
plt.show(block=False)
plt.close()
```
